```python
from requests_oauthlib import OAuth1Session
import json
import datetime, time, sys
import key

session = OAuth1Session(key.CK, key.CS, key.AT, key.AS)
key_words = u'#python'
url = 'https://api.twitter.com/1.1/search/tweets.json'
res = session.get(url, params = {'q':key_words, 'count':30})

if res.status_code != 200:
    print ("Twitter API Error: %d" % res.status_code)
    sys.exit(1)
    
print('„Ç¢„ÇØ„Çª„ÇπÂèØËÉΩÂõûÊï∞ %s' % res.headers['X-Rate-Limit-Remaining'])
print('„É™„Çª„ÉÉ„ÉàÊôÇÈñì %s' % res.headers['X-Rate-Limit-Reset'])
sec =int(res.headers['X-Rate-Limit-Reset'])\
           - time.mktime(datetime.datetime.now().timetuple())
print('„É™„Çª„ÉÉ„ÉàÊôÇÈñì ÔºàÊÆã„ÇäÁßíÊï∞„Å´ÊèõÁÆóÔºâ %s' % sec)
 

res_text = json.loads(res.text)
for tweet in res_text['statuses']:
    print ('--------------------------------------------------------------')
    print('account:',"(fullname:",tweet['user']['name'],'href:',("/"+tweet['user']['screen_name']).strip(),"id:",tweet['user']['id'],')')
    print ("date:",tweet['created_at'])
    print("hashtags:",key_words)
    print(tweet['favorite_count'])
    print(tweet['retweet_count'])
    print (tweet['text'])
```

```python
import tweepy
import key

consumer_key = key.CK
consumer_secret = key.CS
access_key = key.AT
access_secret = key.AS

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_key, access_secret)
api = tweepy.API(auth)
for tweet in tweepy.Cursor(api.user_timeline,screen_name = "@Twitter").items(30):
    print('-------------------------------------------')
    print('full_name:',tweet.user.name,"href:",("/"+tweet.user.screen_name).strip(),"id:",tweet.user.id)
    print("date:",tweet.created_at)
    if tweet.entities['hashtags'] ==[]:
        print('None')
    else:
        print(tweet.entities['hashtags'][0]['text'])
    print(tweet.favorite_count)
    print(tweet.retweet_count)
    print(tweet.text.replace('\n',''))
```
